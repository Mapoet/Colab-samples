```python
import os
import pandas as pd
from google.colab import drive
import json
import base64
drive.mount('/content/drive', force_remount=True)
os.chdir('/content/drive/MyDrive')
import requests
url = "https://www.gnss-x.ac.cn/cgi-bin/colab.py"
params = {"text":base64.b64encode(open('./data/los2stec.cpp','rb').read()).decode('utf8')}
res = requests.get(url=url,params=params)
data=json.loads(res.text)
for line in base64.b64decode(data['args']['text']).decode('utf8').split('\n'):
  print('%s'%line)
```
